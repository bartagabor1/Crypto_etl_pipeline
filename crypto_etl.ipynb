```python
#  Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

#  Generate test data (24 hours * 7 days)
now = datetime.now()
timestamps = [now - timedelta(hours=i) for i in range(7 * 24)]
prices = [60000 + (i % 10 - 5) * 300 for i in range(7 * 24)]  #  simulated price fluctuations

#  Create DataFrame
df = pd.DataFrame({
    "timestamp": timestamps[::-1],
    "price": prices[::-1]
})

#  Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df["timestamp"], df["price"], color='green', linewidth=2)
plt.title("Simulated Bitcoin Price Over Last 7 Days")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()

```
